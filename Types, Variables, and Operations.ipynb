### Problem 6

Operator precedence is the order in which operators activate. Think PEMDAS.

```python
a = 98
b = 36
c = 42
d = 1.9
e = 8**(-1)
f = 12**(-4)
```

```python
f*e**(a+b%d)//c - c**(d/b)
```

Order of execution is as follows:

Parentheses: The values inside the parentheses in the first and second pieces of the function are executed first. The modulus and division occur before addition as in the first piece.

Exponentiation: Exponentiation occurs next.  "e" to the power of the sum of a and the remainder of b mod(d) and "c" to d divided by b.

Multiplication and Division:  f*...//c  occur next.

Subtraction: first piece minus second piece.
